# Feature engineering


```python
import re
df['TouchScreen']="No"
```

```python
#Extra feature "Touchscreen" was extracted (from the screenResolution column) and added to the dataset
#Laptops with touchscreen enabled displays are more costly than those without touchscreen capabilities
#(Personal opinion before EDA)

df.loc[df['ScreenResolution'].str.contains('Touchscreen'),'TouchScreen']="Yes"
df.tail(10)
```

```python
df['ScreenResolution']=df['ScreenResolution'].str.extract('(\d\d\d\d?x\d\d\d\d?)',expand=True)
df.tail(10)
```

```python
#Removal of units : GB,Kg

df['Ram']=df['Ram'].str.extract('(\d\d?)',expand=True)
df['Weight']=df['Weight'].str.extract('(\d\.\d\d?|\d)',expand=True)
df.head(6)
```

Simplification of Cpu column

```python
df['Cpu_copy']=0

#Intel processors
df.loc[df['Cpu'].str.contains('Core M'),'Cpu_copy']="Intel M "
df.loc[df['Cpu'].str.contains('Atom'),'Cpu_copy']="Intel Atom"
df.loc[df['Cpu'].str.contains('Celeron'),'Cpu_copy']="Intel Celeron"
df.loc[df['Cpu'].str.contains('Pentium'),'Cpu_copy']="Intel Pentium"
df.loc[df['Cpu'].str.contains('i3'),'Cpu_copy']="Intel i3"
df.loc[df['Cpu'].str.contains('i5'),'Cpu_copy']="Intel i5"
df.loc[df['Cpu'].str.contains('i7'),'Cpu_copy']="Intel i7"
df.loc[df['Cpu'].str.contains('Xeon'),'Cpu_copy']="Intel Xeon"


#AMD processors
df.loc[df['Cpu'].str.contains('AMD E-Series'),'Cpu_copy']="AMD E"
df.loc[df['Cpu'].str.contains('AMD FX'),'Cpu_copy']="AMD FX"
df.loc[df['Cpu'].str.contains('AMD A4'),'Cpu_copy']="AMD A4"
df.loc[df['Cpu'].str.contains('AMD A6'),'Cpu_copy']="AMD A6"
df.loc[df['Cpu'].str.contains('AMD A8'),'Cpu_copy']="AMD A8"
df.loc[df['Cpu'].str.contains('AMD A9'),'Cpu_copy']="AMD A9"
df.loc[df['Cpu'].str.contains('AMD A10'),'Cpu_copy']="AMD A10"
df.loc[df['Cpu'].str.contains('AMD A12'),'Cpu_copy']="AMD A12"
df.loc[df['Cpu'].str.contains('AMD Ryzen'),'Cpu_copy']="AMD Ryzen"



#Samsung processors
df.loc[df['Cpu'].str.contains('Samsung'),'Cpu_copy']="Samsung"


#print(list(df['Cpu_copy']).count(0))
```

```python
# for i in range(len(df['Cpu_copy'])):
#     if df['Cpu_copy'][i]==0:
#         print(df['Cpu'][i])

df['Cpu']=df['Cpu_copy']
df.drop(columns=['Cpu_copy'],inplace=True)
```

```python
df.head(30)
```

Based on the dataset, laptops have four types of storage medium (SSD , HDD, Flash Storage, SSHD(Hybrid)). Therefore four separate columns/features were created for the each type of memory configuration. The config (SSD + SSD) will be dealt later.

```python
#Dealing with HDD memory


df['HDD']=df['Memory'].str.extract('(\d\d\d?GB\sHDD|\dTB\sHDD|\d\.0TB\sHDD)',expand=True)
df['HDD']=df['HDD'].str.extract('(\d\d\d?|\dTB|\d\.0TB)',expand=True)

df['HDD']=df['HDD'].str.replace('(TB|\.0TB)','000',regex=True)
df['HDD'].fillna(0,inplace=True)

df.head(30)
```

```python
#Used for checking HDD column

# print(df['HDD'].unique())

# for i in range(len(df['HDD'])):
#     if df['HDD'][i]=='0000':
#             print(i)
            
            
```

```python
#Dealing with SSD
df['SSD']=df['Memory'].str.extract('(\d\d?\d?GB\sSSD|\dTB\sSSD|\d\.0TB\sSSD)',expand=True)
df['SSD']=df['SSD'].str.extract('(\dTB|\d\d?\d?|\d\.0TB)',expand=True)

df['SSD']=df['SSD'].str.replace('(TB|\.0TB)','000',regex=True)
df['SSD'].fillna(0,inplace=True)


df.head()
```

Recall that some of the memory configuration consists of 256SSD +256SSD (2 SSD), or multiple SSDs, the total memory of both SSD config would be added. Therefore, this part of the dataset must be fixed manually.

```python
#Shows columns with double SSD config

df.loc[df['Memory'].str.contains('(\d\d\dGB\sSSD\s\+\s\s\d\d\dGB\sSSD)',regex=True)]
```

```python
#Manully combining both SSD config into one

df['SSD'][28]='512'
df['SSD'][266]='1000'
df['SSD'][347]='768'
df['SSD'][901]='512'
```

```python
#Used for checking (as SettingWithCopyWarning given above), https://www.dataquest.io/blog/settingwithcopywarning/

df.loc[df['Memory'].str.contains('(\d\d\dGB\sSSD\s\+\s\s\d\d\dGB\sSSD)',regex=True)]
```

```python
#Dealing with SSHD

df['SSHD']=df['Memory'].str.extract('(\d\d\dGB\sHybrid|\d\.0TB\sHybrid)',expand=True)
df['SSHD']=df['SSHD'].str.extract('(\d\d\d?|\d\.0TB)',expand=True)
df.loc[df['Memory'].str.contains('(\d\d\dGB\sHybrid|\d\.0TB\sHybrid)',regex=True)]
```

```python
df['SSHD']=df['SSHD'].str.replace('(TB|\.0TB)','000',regex=True)
df['SSHD'].fillna(0,inplace=True)
df['SSHD'].unique()
```

```python
#Dealing with Flash Storage
df['FStorage']=df['Memory'].str.extract('(\d\d\d?GB\sFlash)',expand=True)
df['FStorage']=df['FStorage'].str.extract('(\d\d\d?)',expand=True)
df['FStorage'].fillna(0,inplace=True)
print(df['FStorage'].unique())

df.head(10)
```

Gpu column has a high variability of GPUs. Intel Gpus are all integrated GPUS (which have less GPU power than discrete GPUs). Nvidia GPUs are all discrete GPUs. The problem arises when we are analysing AMD branded GPUs. Some AMD GPUs are integrated and some are discrete. Labelling each AMD GPU would take significant effort and time which may or may not contribute to the modelling proccess later on. Due to this issue, I've decided to drop the entire GPU column for simplicity purposes.

```python
df.drop(columns=['Gpu'],inplace=True)
df
```
